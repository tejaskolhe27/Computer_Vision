```python
import matplotlib.pyplot as plt 
import numpy as np 
from PIL import Image
import cv2
```

```python
img = cv2.imread("images/dog.jpg")
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

- Precedence: Resize factors (fx, fy) take precedence over the output size tuple ((0, 0)) when both are provided.
- Aspect ratio: Even when (0, 0) is used, the function generally preserves the aspect ratio of the image unless you explicitly change it with additional parameters.

```python
##cv2.image(image,,fx=percentage,fy=percentage)
newImg  = cv2.resize(img,(0,0),fx = 0.75,fy = 0.75)
```

```python
newImg.shape
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img)

plt.subplot(1,2,2)
plt.title("Cropped Image")
plt.imshow(newImg)
plt.show()
```

```python
##cv2.image(image,,fx=percentage,fy=percentage)
newImg  = cv2.resize(img,(0,0),fx = 1.25,fy = 1.25)

plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img)

plt.subplot(1,2,2)
plt.title("Cropped Image")
plt.imshow(newImg)
plt.show()
```
