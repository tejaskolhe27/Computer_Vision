```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img = cv2.imread("images/draw.png")
plt.imshow(img);
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = np.uint8(img)
```

```python
# perofrm edge detection (low threshold,high threshold)
edges = cv2.Canny(img,50,150)
```

```python
plt.imshow(img)
```

```python
plt.imshow(edges,cmap= 'gray')
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img)

plt.subplot(1,2,2)
plt.title("Edged Image")
plt.imshow(edges,cmap= 'gray')
plt.show()
```

## Laplacian Edge Detection

```python
img = cv2.imread("images/houseleft.jpg")
```

```python
#perform Laplacian edge detection
kernel = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])

#src,ddeth(-1 ignore),kernel
edges = cv2.filter2D(img,-1,kernel)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img)

plt.subplot(1,2,2)
plt.title("Edged Image")
plt.imshow(edges,cmap= 'gray')
plt.show()
```
