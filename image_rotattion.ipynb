```python
import matplotlib.pyplot as plt 
import numpy as np 
from PIL import Image
import cv2
```

```python
img = cv2.imread("images/caves.jpg")
```

```python
img.shape
```

```python
img.transpose().shape
```

```python
new_img = img.reshape(400,534,3)
```

```python
new_img.shape
plt.imshow(new_img)
```

```python
#converting BGR to RGB Format
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
## cv2.getRotationMatrix2D(center,angle,scale)
width,height,_ = img.shape
rotationMatrix = cv2.getRotationMatrix2D((width/2,height/2),45,1)
```

```python
rotationMatrix.shape
```

```python
rotatedImage = cv2.warpAffine(img,rotationMatrix,(width,height))
```

```python
plt.imshow(rotatedImage)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.title("Original Image")
plt.imshow(img)

plt.subplot(1,2,2)
plt.title("Cropped Image")
plt.imshow(rotatedImage)
plt.show()
```
