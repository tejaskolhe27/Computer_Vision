```python
import matplotlib.pyplot as plt
import numpy as np 
from PIL import Image
import cv2
```

```python
img  = cv2.imread("images/caves.jpg")
```

```python
plt.imshow(img);
```

```python
plt.imshow(img[100:300,100:300])
```

```python
img.shape
```

```python
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img);
```

```python
plt.imshow(img[80:454,60:340])
```

```python
height, width,_ = img.shape
```

```python
height*(15/100)
```

```python
new_img = rgb_img[int(height*(15/100)):(height - int(height*(15/100))),int(width*(15/100)) : (width - int(width*(15/100)))]
```

```python
plt.imshow(new_img)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.title("Original Image")
plt.imshow(rgb_img)

plt.subplot(1,2,2)
plt.title("Cropped Image")
plt.imshow(new_img)
plt.show()
```

```python
img = Image.open("images/dog.jpg")
img
```

```python
height,width = img.size
```

```python
first = img.crop((0,0,int(height/2),int(width/2)))
second = img.crop((int(width/2),0,width,int(height/2)))
third = img.crop((0,int(height/2),int(width/2),height))
forth = img.crop((int(width/2),int(height/2),width,height))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and a grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Add images to the subplots
axs[0, 0].imshow(first)
axs[0, 1].imshow(second)
axs[1, 0].imshow(third)
axs[1, 1].imshow(forth)

# Set the titles for the subplots
axs[0, 0].set_title('Image 1')
axs[0, 1].set_title('Image 2')
axs[1, 0].set_title('Image 3')
axs[1, 1].set_title('Image 4')


axs[0, 0].grid()
axs[0, 1].grid()
axs[1, 0].grid()
axs[1, 1].grid()

# Show the plot
plt.show()
```
